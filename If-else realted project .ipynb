```python
# Student Result Management System

n = int(input("Enter number of students: "))

for i in range(n):
    marks = int(input("Enter marks: "))

    if marks >= 0 and marks <= 100:
        if marks >= 90:
            grade = "A"
        elif marks >= 75:
            grade = "B"
        elif marks >= 50:
            grade = "C"
        else:
            grade = "Fail"
        print("Grade:", grade)
    else:
        print("Invalid marks")
```

```python
n = int(input("enter student númer: "))

for i in range(n):
  name = input("Enter student name: ")
  marks1 = int(input("Enter marks: "))
  marks2 = int(input("Enter marks: "))
  marks =  int(input("Enter marks: "))

  percentage = (marks1 + marks2 + marks)/3
  percentage = round(percentage, 2)
  print(f"percentage of student is { percentage}")

  if percentage >= 90 and percentage <=100:
    print(f"Name: {name} get  Grade: A")
  elif percentage >= 70 and percentage <=90:
    print(f"Name: {name} get  Grade: B")
  elif percentage >= 50 and percentage <=70:
    print(f"Name: {name} get  Grade: C")
  else:
    print(f"Name: {name} get  Grade: Fail")

  print("------------------------------")
```
