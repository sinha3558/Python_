```python
class student:
  def __init__(self,name,marks):
    self.name = name
    self.mrks = marks
    print("add new student in class: ")

  def get_avg(self):
    sum = 0
    for i in self.mrks:
      sum = sum + i
    print(f"welcome to {self.name} and average marks is {round(sum/3,2)} % in board exam")

s2 = student("Raj",[90,89,96])
print(s2.name)
print(s2.mrks)
print(s2.get_avg())
```

```python
class account:
  def __init__(self,name,balance,acc_no):
    self.name = name
    self.balance = balance
    self.acc_no = acc_no
    print("welcome to bank")

  def debit(self,debit_amount):
      self.balance -= debit_amount
      print("Rs", debit_amount)
      print(f'my final total balnce is {self.get_balance()}')


  def credit(self,credit_amount):
      self.balance += credit_amount
      print("Rs", credit_amount)
      print(f'my final total balnce is {self.get_balance()}')


  def get_balance(self):
      return self.balance

name = input("enter your name: ")
acc_no = int(input("enter your account no: "))
balance = int(input("enter your balance: "))
debit_amount = int(input("enter your debit amount: "))
credit_amount = int(input("enter your credit amount: "))


account2 = account(name,balance,acc_no)
print(account2.name)
print(account2.balance)
print(account2.acc_no)
account2.debit(debit_amount)
account2.credit(credit_amount)

print("total balance", account2.get_balance())
```
