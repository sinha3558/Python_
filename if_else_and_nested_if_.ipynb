```python
# Leap year
#A year is a leap year if:Year 400 से divisible हो OR Year 4 से divisible हो और 100 से divisible न हो

year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")

```
